# Simple Linear Regression

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
df=pd.read_csv('Salary_Data.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression() #regressor is an object of class LinearRegression
regressor.fit(X_train,y_train)
```

## Predicting the Test set results

```python
y_pred=regressor.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train,y_train,color='red')
plt.show()
```

```python
pred_on_training=regressor.predict(X_train)
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,pred_on_training,color='blue')
plt.title('Salary vs Experience(Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test,y_test,color="red")
plt.plot(X_test,y_pred,color="blue") #We could also write plt.plot(X_train,pred_on_training,color="blue") instead. Both will result in the same line
plt.title('Salary vs Experience(Test Set)')
plt.xlabel('Salary')
plt.ylabel('Experience')
plt.show()
```
